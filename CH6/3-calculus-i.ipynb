Selanjutnya kita akan mengimplementasikan Limit pada Non Continuous Function
$$\lim_{x \to 1} \frac{x^2 - 1}{x - 1}$$

```python
# Kita panggil library "matplotlib" untuk visualisasi

import matplotlib.pyplot as plt
```

```python
# Import NumPy library
# NumPy merupakan library yang digunakan untuk memproses data berbentuk Array
# Library ini juga digunakan untuk kasus linear algebra, fourier transform, dan matrices

import numpy as np
```

```python
# Generate sebuah vektor menggunakan NumPy

x = np.linspace(-10, 10, 10000)
x
```

```python
# Tulis persamaan di atas menggunakan Python
# Sebagai berikut:

def my_fxn(my_x):
    my_y = (my_x**2 - 1)/(my_x - 1)
    return my_y
```

```python
# Operasikan fungsi yang sudah kita buat dengan memasukkan nilai "x"
# Setelah itu akan muncul hasil dari fungsi 
# Sebagai berikut

my_fxn(2)
```

```python
# Jika fungsi diisi nilai "1", maka akan menghasilkan 'division by zero'
# Karena jika memasukkan nilai "1", akan menimbulkan kondisi "0" dibagi "0"
my_fxn(1)
```

```python
# Kita masukan vektor yang sudah tergenerate tadi ke fungsi

y = my_fxn(x)
```

```python
# Visualisasikan kasus Limit pada Non Continuous Function
# Terlihat ketika nilai semakit dekat dengan "1",
# Maka ia memiliki limit "2"

fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.axvline(x=1, color='purple', linestyle='--')
plt.axhline(y=2, color='purple', linestyle='--')
_ = ax.plot(x,y)
```

Selanjutnya kita akan mengimplementasikan Limit pada Infinity
$$ \lim_{x \to \infty} \frac{25}{x} $$

### Delta Method

```python
# Kita panggil library "matplotlib" untuk visualisasi

import matplotlib.pyplot as plt
```

```python
# Import NumPy library
# NumPy merupakan library yang digunakan untuk memproses data berbentuk Array
# Library ini juga digunakan untuk kasus linear algebra, fourier transform, dan matrices

import numpy as np
```

```python
# Generate sebuah vektor menggunakan NumPy

x = np.linspace(-10, 10, 10000)
```

Selanjutnya tulis persamaan berikut ke dalam Python. 
Persamaan ini sudah kita pelajari pada sesi sebelumnya:
$y = x^2 + 2x + 2$:

```python
def f(my_x):
    my_y = my_x**2 + 2*my_x + 2
    return my_y
```

```python
# Kita masukan vektor yang sudah tergenerate tadi ke fungsi

y = f(x)
```

```python
# Visualisasikan dan akan tampil visualisasi seperti ini

fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
_ = ax.plot(x,y)
```

Selanjutnya kita akan cek slope, dimana $x = 2$.

Pertama, kita tentukan $y$: 

```python
# Maka kita akan mendapatkan nilai y adalah "10"

f(2)
```

Selanjutnya kita sebut x dan y dengan titik $P$, yang mana pada (2, 10):

```python
# Visualisasikan titik di atas, 
# sehingga kita bisa melihat posisi sebenarnya dari titik (2, 10).

fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.scatter(2, 10) # new
_ = ax.plot(x,y)
```

_Delta method_ memerlukan 2 titik untuk menghitung slope. Untuk mempraktekan hal berikut, mari kita tentukan poin lainnya, yakni titik $Q$, dimana $x = 5$.

```python
# Sama seperti sebelumnya kita perlu mencari nilai "y" dari nilai "x = 5"
# Maka kita akan mendapatkan nilai "y = 37"

f(5)
```

```python
# Visualisasikan titik (5, 37), sehingga kita bisa melihat posisi sebenarnya

fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.scatter(2, 10)
plt.scatter(5, 37, c = 'orange', zorder=3) # new
_ = ax.plot(x,y)
```

Selanjutnya kita menggunakan persamaan berikut untuk mencari slope $m$ (notasi untuk slope) diantara titik $P$ dan $Q$, dimana: 


*   $x_1$ => 2
*   $y_1$ => 10
*   $x_2$ => 5
*   $y_2$ => 37

Hasilnya nilai $m$ adalah 9


$$m = \frac{\text{change in }y}{\text{change in }x} = \frac{\Delta y}{\Delta x} = \frac{y_2 - y_1}{x_2 - x_1} = \frac{37-10}{5-2} = \frac{27}{3} = 9$$

```python
# Implementasikan persamaan di atas pada Python
# Maka kita akan mendapatkan nilai slope "9"

m = (37-10)/(5-2)
m
```

Untuk melakukan visualisasi garis yang melalui $P$ dan $Q$, kita dapat mengatur ulang persamaan garis $y = mx + b$ untuk agar mendapatkan nilai $b$, menjadi:
$$b = y - mx$$

```python
# Kita hitung nilai "b" menggunakan Python, dimana:
# "y" menggunakan nilai "y2", yakni "37"
# "m" menggunakan nilai yang sudah kita hitung yakni "9"
# "x" menggunakan nilai "x2", yakni "5" 

b = 37-m*5
b
```

Selanjutnya kita bisa menghitung persamaan $y = mx + b$

```python
line_y = m*x + b
```

```python
# Lalu visualisasikan, menjadi sebagai berikut:

fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
# Poin/titik "p" berwarna "biru"
plt.scatter(2, 10)
# Poin/titik "q" berwarna "oranye"
plt.scatter(5, 37, c='orange', zorder=3)
plt.ylim(-5, 150)
# Garis "oranye" yang mengacu ke persamaan "m*x + b"
plt.plot(x, line_y, c='orange')
_ = ax.plot(x,y)
```

Semakin dekat nilai $Q$ mendekati $P$, maka akan semakin dekat slope $m$ menjadi "true tangent" dari titik $P$. 

Baiklah, mari kita coba mengganti nilai x untuk titik menjadi 2.1$.

Sama seperti sebelumnya, jika $\Delta x$ antara $Q$ dan $P$ sama dengan 3, maka sekarang lebih kecil menjadi: $$\Delta x = x_2 - x_1 = 2.1 - 2 = 0.1 $$

Kita akan 
mencoba mencari delta method untuk titik P guna mencari slope, dimana $x = -1$.

```python
# Kita panggil library "matplotlib" untuk visualisasi

import matplotlib.pyplot as plt
```

```python
# Import NumPy library
# NumPy merupakan library yang digunakan untuk memproses data berbentuk Array
# Library ini juga digunakan untuk kasus linear algebra, fourier transform, dan matrices

import numpy as np
```

```python
# Generate sebuah vektor menggunakan NumPy

x = np.linspace(-10, 10, 10000)
```

Selanjutnya tulis persamaan berikut ke dalam Python. 
Persamaan ini sudah kita pelajari pada sesi sebelumnya:
$y = x^2 + 2x + 2$:

```python
def f(my_x):
    my_y = my_x**2 + 2*my_x + 2
    return my_y
```

```python
# Kita masukan vektor yang sudah tergenerate tadi ke fungsi

y = f(x)
```

```python
# Definisikan variabel x1 = -1

x1 = -1
```

```python
# Cari nilai "y1" dari titik "P" dimana nilai "x1 = -1",
# sehingga mendapatkan nilai "y = 1"

y1 = f(x1)
y1
```

Maka titik $P$ berlokasi pada (-1, 1)

```python
# Kita definisikan nilai "delta x" sebesar "0.000001",

delta_x = 0.000001
```

```python
# Buat titik "q", dimana memiliki "x2" yang didapat dari "x1" + "delta_x"
# sehingga menghasilkan nilai "-0.999999"

x2 = x1 + delta_x
x2
```

```python
# Cari nilai "y2" dari titik "q" dimana nilai "x2 = -0.999999",
# sehingga mendapatkan nilai "y2 = 1.000000000001"

y2 = f(x2)
y2
```

Maka kita mendapatkan titik $Q$ pada (-0.999999, 1.000000000001), sehingga bisa dikatakan mendekati titik $P$.

```python
# Selanjutnya kita hitung slope nya,
# hasilnya nilai slope nya adalah "1.0000889005535828e-06"

m = (y2-y1)/(x2-x1)
m
```

Oleh karena itu, karena $x_2$ menjadi sangat dekat dengan $x_1$, menjadi jelas bahwa slope $m$ pada $x_1 = -1$ sama dengan nol. Setelah itu mari kita visualisasikan:

```python
# Terlebih dahulu kita ambil hitung nilai "b" nya
# Sehingga bisa mendapatkan nilai "1.0000010000889006"

b = y2-m*x2
b
```

Lalu kita bisa menghitung persamaan $y = mx + b$

```python
line_y = m*x + b
```

```python
# Lalu visualisasikan, menjadi sebagai berikut:

fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
# Poin/titik "p" berwarna "biru"
plt.scatter(x1, y1)
# Poin/titik "q" (dengan nilai baru) berwarna "oranye"
plt.scatter(x2, y2, c='orange', zorder=3)
plt.ylim(-5, 150)
# Garis "oranye" yang mengacu ke persamaan "m*x + b"
plt.plot(x, line_y, c='orange', zorder=3)
_ = ax.plot(x,y)

# Maka akan menghadilkan visualisasi dimana titik "p" semakin dekat dengan titik "q"
# dan diikuti juga slope yang semakin dekat dengan titik "q"
```

Dari pembuktian di atas titik $Q$ dekat dengan titik $P$, bisa dikatakan:
* $x_2$ - $x_1$ mendekati 0
* Dengan kata lain, $\Delta x$ mendekati 0
* Hal ini bisa dinotasikan dengan $\Delta x \to 0$

Berdasarkan fakta di atas derivative dari $y$ (dinotasikan $dy$) terhadap $x$ (dinotasikan $dx$) dapat ditulis dengan persamaan: 
$$\frac{dy}{dx} = \lim_{\Delta x \to 0} \frac{\Delta y}{\Delta x}$$

$\Delta y$ dapat dijabarkan menjadi $y_2 - y_1$: 
$$\frac{dy}{dx} = \lim_{\Delta x \to 0} \frac{y_2 - y_1}{\Delta x}$$

Kemudian replace $y_1$ dengan $f(x)$ dan replace $y_2$ dengan $f(x + \Delta x)$, sehingga kita mendapatkan persamaan sebagai berikut:
$$\frac{dy}{dx} = \lim_{\Delta x \to 0} \frac{f(x + \Delta x) - f(x)}{\Delta x}$$

Selanjutnya kita akan menerapkan persamaan di atas pada Python: 

```python
def diff_demo(my_f, my_x, my_delta):
    return (my_f(my_x + my_delta) - my_f(my_x)) / my_delta
```

```python
# Kita definisikan nilai delta sebagai berikut
# Di bawah ini terdapat nilai delta dari yang besar sampai kecil

deltas = [1, 0.1, 0.01, 0.001, 0.0001, 0.00001, 0.000001]
```

```python
# Lalu kita cek masing-masing delta pada fungsi "diff_demo",
# dan menghasilkan nilai sebagai berikut

for delta in deltas:
    print(diff_demo(f, 2, delta))
```

```python
for delta in deltas:
    print(diff_demo(f, -1, delta))
```

## Automatic Differentiation

**TensorFlow** merupakan salah satu algoritma yang populer untuk melakukan automatic differentiation. Berikut kita akan mencoba mempraktekannya.

Mari kita coba gunakan Tensorflow untuk mengkalkulasikan $dy/dx$ pada $x = 5$, dimana: 

$$y = x^2$$

$$ \frac{dy}{dx} = 2x = 2(5) = 10 $$

### Autodiff with TensorFlow

```python
# Import library Tensorflow

import tensorflow as tf
```

```python
# Definisikan variabel "x" yang didalamnya terdapat nilai scalar "5.0"

x = tf.Variable(5.0)
```

```python
# Lacak gradient melalui forward pass

with tf.GradientTape() as t:
    t.watch(x)
    y = x**2
```

```python
# Ambil nilai slope menggunakan fitur autodiff dari Tensorflow
# Sehingga kita mendapatkan nilai "10"

t.gradient(y, x) # use autodiff
```
