# Creating different types of graphs with Pyplot

**Bar Graphs**

```python
divisions = ["Div A", "Div B", "Div C", "Div D", "Div E"]
division_average_marks = [70,  82, 73, 65, 68]

plt.bar(divisions, division_average_marks, color='green')
plt.title("Bar Graph")
plt.xlabel("Divisions")
plt.ylabel("Marks")
plt.show()
```

```python
divisions = ["Div A", "Div B", "Div C", "Div D", "Div E"]
division_average_marks = [70,  82, 73, 65, 68]
variance = [5,8,7,6,4]

plt.barh(divisions, division_average_marks, xerr=variance, color='green')
plt.title("Bar Graph")
plt.xlabel("Marks")
plt.ylabel("Divisions")
plt.show()
```

```python
divisions = ["Div A", "Div B", "Div C", "Div D", "Div E"]
division_average_marks = [70,  82, 73, 65, 68]
boys_average_marks = [68, 67, 77, 61, 70]

index = np.arange(5) # [0, 1, 2, 3, 4]
width = 0.30

plt.bar(index, division_average_marks, width, color='green', label='Division Marks')
plt.bar(index+width, boys_average_marks, width, color='blue', label='Boys Marks')
plt.title("Horizontally Stacked Bar Graphs")

plt.ylabel("Marks")
plt.xlabel("Divisions")
plt.xticks(index+ width/2, divisions)

plt.legend(loc='best')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

divisions = ["Div A", "Div B", "Div C", "Div D", "Div E"]
division_average_marks = [70, 82, 73, 65, 68]
boys_average_marks = [68, 67, 77, 61, 70]

index = np.arange(5) # [0, 1, 2, 3, 4]
width = 0.30

plt.bar(index, division_average_marks, width, color='green', label='Division Marks')
plt.bar(index+width, boys_average_marks, width, color='blue', label='Boys Marks')
plt.title("Horizontally Stacked Bar Graphs")

plt.ylabel("Marks")
plt.xlabel("Divisions")
plt.xticks(index + width / 2, divisions)

plt.legend(loc='best')

# Menambahkan label angka di dalam stacked bar
for i in range(len(divisions)):
    plt.text(index[i] + width / 2, division_average_marks[i] + 2, division_average_marks[i], ha='center', va='bottom', color='black')
    plt.text(index[i] + width * 1.5, boys_average_marks[i] + 2, boys_average_marks[i], ha='center', va='bottom', color='black')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

divisions = ["Div A", "Div B", "Div C", "Div D", "Div E"]
division_average_marks = [70, 82, 73, 65, 68]
boys_average_marks = [68, 67, 77, 61, 70]

index = np.arange(5) # [0, 1, 2, 3, 4]
width = 0.30

plt.bar(index, division_average_marks, width, color='blue', label='Division Marks')
plt.bar(index, boys_average_marks, width, bottom=division_average_marks, color='red', label='Boys Marks')
plt.title("Vertically Stacked Bar Graphs")

plt.ylabel("Marks")
plt.xlabel("Divisions")
plt.xticks(index, divisions)

plt.legend(loc='best')

# Menambahkan label angka di dalam stacked bar
for i in range(len(divisions)):
    plt.text(index[i], division_average_marks[i] / 2, division_average_marks[i], ha='center', va='center', color='white')
    plt.text(index[i], division_average_marks[i] + boys_average_marks[i] / 2, boys_average_marks[i], ha='center', va='center', color='white')

plt.show()
```

**Pie Charts**

```python
firms = ["Firm A", "Firm B", "Firm C", "Firm D", "Firm E"]
market_share = [20, 25, 15, 10, 20]
Explode = [0,0.1,0,0,0]
plt.pie(market_share,explode=Explode,labels=firms,shadow=True,
        startangle=45)
plt.axis('equal')
plt.legend(title="List of Firms")
plt.show()
```

```python
x = np.random.randn(1000)

plt.title("Histogram")
plt.xlabel("Random Data")
plt.ylabel("Frequency")
plt.hist(x,10)
plt.show()
```

Batas ini biasanya dipilih berdasarkan kebutuhan analisis dan preferensi visual agar bisa memahami sebaran data dan pola distribusi dengan cara yang lebih jelas di area tertentu

```python
height = np.array([167,170,149,165,155,180,166,146,
                   159,185,145,168,172,181,169])

weight = np.array([86,74,66,78,68,79,90,73,
                   70,88,66,84,67,84,77])

plt.xlim(140,200)
plt.ylim(60,100)
plt.scatter(height,weight)
plt.title("Scatter Plot")
plt.xlabel("Height")
plt.ylabel("Weight")
plt.show()
```
