# Seaborn's plotting functions

```python
# Recommended way
sns.lmplot(x='Attack', y='Defense', data=df)
 
# Alternative way, sudah outdated
# sns.lmplot(x=dfAttack, y=df.Defense)
```

```python
sns.lmplot(x='Attack', y='Defense', data=df,
           fit_reg=False, 
           hue='Stage')   
```

# Customizing with Matplotlib

```python
# Plot using Seaborn
sns.lmplot(x='Attack', y='Defense', data=df,
           fit_reg=False, 
           hue='Stage')
 
# Tweak using Matplotlib
# parameter (ymin, ymax) & (xmin, xmax)
plt.ylim(0, None)
plt.xlim(0, None)
```

# The role of Pandas

```python
# Boxplot
sns.boxplot(data=df)
```

```python
# Pre-format DataFrame
stats_df = df.drop(['Total', 'Stage', 'Legendary'], axis=1)
 
# New boxplot using stats_df
sns.boxplot(data=stats_df)
```

# Heatmap

```python
stats_df = df[['HP', 'Attack',	'Defense', 'Sp. Atk', 'Sp. Def','Speed']]
corr = stats_df.corr()
 
sns.heatmap(corr)
```

# Histogram

```python
# distribution plot
sns.distplot(df.Attack)
```

# Bar Plot

```python
# Count Plot (a.k.a. Bar Plot)
sns.countplot(x='Type 1', data=df)
 
# Rotate x-labels
plt.xticks(rotation=-45)
```

# Factor Plot

```python
# Factor Plot --> sudah outdated ganti dengan catplot
# catplot: untuk memvisualisasi data yang sifatnya kategori
g = sns.catplot(x='Type 1', 
                   y='Attack', 
                   data=df, 
                   hue='Stage',  # Color by stage
                   col='Stage',  # Separate by stage
                   kind='swarm') # Swarmplot
 
# Rotate x-axis labels
g.set_xticklabels(rotation=-45)
 
# Doesn't work because only rotates last plot
# plt.xticks(rotation=-45)
```

biasanya dipakai untuk melihat dan memahami sebaran, perbedaan, dan hubungan antar variabel-variabel.

# Density Plot

```python
# Density Plot KDE (Kernel Density Estimation)

# sns.kdeplot(df.Attack, df.Defense) --> outdated
sns.kdeplot(x='Attack', y='Defense', data=df)
```

Tambahan informasi

melihat apakah Distribusi dari suatu data bersifat simetris atau tidak

- simetris: punya nilai rata-rata yang terletak di pusat distribusi, dan proporsi data di sebelah kiri dan kanan rata-rata cenderung mirip / serupa

- asimetris: sebelah kiri atau kanan rata-rata distribusi lebih tebal (lebih banyak data) dibandingkan sisi yang lain. Distribusi asimetris dibagi jadi dua jenis: positif (skewness positif) dan negatif (skewness negatif). Ciri2 distribusi positif, ekor panjang lebih ke kanan, sedangkan dalam distribusi negatif, ekor panjang lebih ke kiri. akan dibahas lebih lanjut pakai histogram
