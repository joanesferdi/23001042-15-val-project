```python
import yfinance as yf

# Define the ticker symbols for KO and PEP
# tickers = ["KO", "PEP"]
# tickers = ['MSFT', 'AAPL']
# tickers = ['V', 'MA']
# tickers = ['PG', 'JNJ']
# tickers = ['T', 'VZ']
tickers = ['TLKM.JK', 'BBCA.JK']

# Retrieve the historical prices for the tickers
prices = yf.download(tickers, start="2023-01-01", end="2024-26-01")["Adj Close"]

# Print the last 5 rows of the prices DataFrame
prices.to_csv('stock_data.csv')
```

```python
data = pd.read_csv('stock_data.csv')
```

```python
data.head()
```

```python
# Rename kolom
data = data.rename(columns={'BBCA.JK': 'BBCA', 'TLKM.JK': 'TLKM'})
```

```python
# Menampilkan statistik sederhana
print("Statistik Saham BBCA:")
print(data['BBCA'].describe())
print("\nStatistik Saham TLKM:")
print(data['TLKM'].describe())

# Menampilkan histogram untuk harga saham BBCA
plt.hist(data['BBCA'], bins=10, alpha=0.6, color='b', label='BBCA')
plt.xlabel('Harga Saham')
plt.ylabel('Frekuensi')
plt.title('Distribusi Harga Saham BBCA')
plt.legend()
plt.show()

# Menampilkan histogram untuk harga saham TLKM
plt.hist(data['TLKM'], bins=10, alpha=0.6, color='r', label='TLKM')
plt.xlabel('Harga Saham')
plt.ylabel('Frekuensi')
plt.title('Distribusi Harga Saham TLKM')
plt.legend()
plt.show()
```

```python
# Menampilkan histogram untuk harga saham BBCA
plt.hist(data['BBCA'], color='lightgray')

# Menambahkan garis vertikal untuk rata-rata dan median harga saham BBCA
plt.axvline(x=np.mean(data['BBCA']), color='orange', label='Mean')
plt.axvline(x=np.median(data['BBCA']), color='green', label='Median')

plt.xlabel('Harga Saham')
plt.ylabel('Frekuensi')
plt.title('Distribusi Harga Saham BBCA')
plt.legend()
plt.show()

# Menampilkan histogram untuk harga saham TLKM
plt.hist(data['TLKM'], color='lightgray')

# Menambahkan garis vertikal untuk rata-rata dan median harga saham TLKM
plt.axvline(x=np.mean(data['TLKM']), color='orange', label='Mean')
plt.axvline(x=np.median(data['TLKM']), color='green', label='Median')

plt.xlabel('Harga Saham')
plt.ylabel('Frekuensi')
plt.title('Distribusi Harga Saham TLKM')
plt.legend()
plt.show()
```
