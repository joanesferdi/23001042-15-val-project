```python
import pandas as pd
from scipy.stats import f_oneway
```

```python
df = pd.read_csv('dataset_phishing.csv')
```

```python
df
```

```python
df.info()
```

```python
# List untuk menyimpan p-value dari setiap fitur
p_values = []

# looping melalui setiap kolom fitur kecuali kolom 'url' dan 'status'
for column in df.columns[1:-1]:
    f_statistic, p_value = f_oneway(df[df['status'] == 'phishing'][column],
                                     df[df['status'] == 'legitimate'][column])
    p_values.append((column, p_value))

# CARA PERTAMA
# def sort_by_second_element(x):
#     return x[1]
# p_values.sort(key=sort_by_second_element)

# CARA KEDUA
# Sort p-value dari yang terkecil
p_values.sort(key=lambda x: x[1])

print("Hasil Analisis ANOVA:")
print("=====================")
for feature, p_value in p_values:
    print(f"Nama Fitur: {feature}, Nilai p-value: {p_value}")

# Fitur terbaik adalah yang memiliki p-value terkecil
best_feature, best_p_value = p_values[0]
print("Fitur Terbaik:")
print("================")
print(f"Nama Fitur: {best_feature}")
print(f"Nilai p-value: {best_p_value}")
```

```python
# Buat DataFrame dari p_values
output_df = pd.DataFrame(p_values, columns=['Feature', 'P-Value'])

# Urutkan DataFrame berdasarkan nilai p-value terkecil
output_df = output_df.sort_values(by='P-Value')

# Reset index setelah pengurutan
output_df.reset_index(drop=True, inplace=True)

# Tampilkan DataFrame yang sudah diurutkan
output_df
```

p-value < alpha (0.05): Jika nilai p-value lebih kecil dari alpha (tingkat signifikansi yang ditentukan sebelumnya, umumnya 0.05), maka terdapat cukup bukti statistik untuk menolak hipotesis nol. Ini berarti bahwa perbedaan atau pengaruh variabel independen (x) terhadap variabel dependen (y) adalah signifikan secara statistik.

p-value >= alpha: Jika nilai p-value lebih besar atau sama dengan alpha, maka tidak cukup bukti statistik untuk menolak hipotesis nol. Ini berarti bahwa tidak terdapat cukup bukti untuk menunjukkan adanya perbedaan atau pengaruh yang signifikan secara statistik antara variabel independen dan dependen.

Dalam konteks ini, nilai p-value yang mendekati nol menunjukkan bahwa terdapat bukti statistik yang sangat kuat untuk menolak hipotesis nol, yang berarti bahwa variabel-variabel yang terkait dengan fitur memiliki pengaruh yang signifikan terhadap klasifikasi phishing atau tidak phishing. Sedangkan nilai p-value yang bernilai 0 menunjukkan bahwa perbedaan atau pengaruh tersebut sangat signifikan secara statistik.
