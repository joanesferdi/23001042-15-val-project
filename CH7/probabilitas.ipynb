```python
ns = np.array([2, 4, 8, 32, 64, 128, 256, 512, 1024, 2048, 4096])
```

```python
head_count = [np.random.binomial(n, 0.5) for n in ns]
head_count
```

```python
for n in ns:
    print(np.random.binomial(n, 0.5))
```

```python
proprtion_heads = head_count/ns
proprtion_heads
```

```python
fig, ax = plt.subplots()
plt.xlabel('Number of coin flips in experiments')
plt.ylabel('Proportion of flips that are heads')
plt.axhline(0.5, color='orange')
_ = ax.scatter(ns, proprtion_heads)
```

```python
n_experiment = 1000
head_count = np.random.binomial(5, 0.5, n_experiment)
```

```python
heads, event_count = np.unique(head_count, return_counts=True)
```

```python
heads
```

```python
event_count
```

```python
event_proba = event_count / n_experiment
event_proba
```

```python
plt.bar(heads, event_proba, color=['red', 'yellow', 'black', 'blue', 'orange', '#7eb54e'])
plt.xlabel('Heads flips (out of 5 tosses)')
_ = plt.ylabel('Event probability')
```
